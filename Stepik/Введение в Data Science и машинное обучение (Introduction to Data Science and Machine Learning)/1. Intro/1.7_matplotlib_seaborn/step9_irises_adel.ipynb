```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
irises = pd.read_csv('../data/iris.csv', index_col=0)
```

```python
irises.head()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0] = sns.distplot(irises['sepal length'], ax=axes[0, 0])
axes[0, 1] = sns.distplot(irises['sepal width'], ax=axes[0, 1])
axes[1, 0] = sns.distplot(irises['petal length'], ax=axes[1, 0])
axes[1, 1] = sns.distplot(irises['petal width'], ax=axes[1, 1])
plt.subplots_adjust(bottom=8, top=10, left=7, right=8)
```

### Step 2

```python
students_performance.groupby('gender').mean()
```

```python
students_performance.groupby('gender').aggregate(
    mean_math_score=('math score', 'mean'),
    mean_reading_score=('reading score', 'mean'))
```

### Step 3

```python
students_performance.groupby(['gender', 'race/ethnicity']).aggregate(
    mean_math_score=('math score', 'mean'),
    mean_reading_score=('reading score', 'mean'))
```

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity']).aggregate(
    mean_math_score=('math score', 'mean'),
    mean_reading_score=('reading score', 'mean'))
```

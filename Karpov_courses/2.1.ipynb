7. Сохраните название района с наименьшим числом поездок в переменную min_pickups.

```python
min_pickups = taxi.groupby('borough') \
    .aggregate({'pickups': 'sum'}) \
    .idxmin()
```

8. Сгруппируйте данные по двум признакам: району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и назовите районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни. 

```python
taxi.groupby(['borough', 'hday']) \
    .aggregate({'pickups': 'mean'})
```

9. Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и сохраните результирующий датафрейм в pickups_by_mon_bor. Обратите внимание, что итоговый датасет должен состоять из 3-х колонок – pickup_month, borough, pickups.

```python
pickups_by_mon_bor = taxi.groupby(['borough', 'pickup_month'], as_index=False) \
    .aggregate({'pickups': 'sum'}) \
    .sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor
```

10. Поскольку данные о поездках в Нью-Йорке, температура представлена в градусах Фаренгейта. Напишите функцию temp_to_celcius, которая получает на вход колонку с температурой в °F, и возвращает значения, переведенные в градусы Цельсия.

```python
def temp_to_celcius(temp):
    return ((temp-32) * 5.0)/(9.0)
```

```python
temp_to_celcius(taxi.temp)
```
